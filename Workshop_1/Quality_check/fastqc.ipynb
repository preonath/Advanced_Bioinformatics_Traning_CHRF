## **FastQC** ##


>- **A typical whole genome sequencing process involves**
>>- **sample preparation,** 
>>- **library preparation** 
>>- **and sequencing.** 
>- **Errors occurring at each of these steps can negatively impact the quality of the sequence information.** 
>- **It is important to perform quality checks on the raw sequence reads before starting your analysis.**

### **Some of the parameters for NGS data quality control.** ###
>- **Check for primers and adaptors**
>- **Trim low quality ends**
>- **Remove low quality reads**
>- **Remove short sequences**
>- **Remove reads with a high proportion of ambiguous bases (“N”)**
>- **Remove duplicates**



## **Tool** ##

**We will be using the FastQC tool to carry out quality control on reads.** 


## **Installation:** ## 

``` bash
conda install -c bioconda fastqc
```

### **Input files:** ###

>- **To run FastQC,**
>>- **you need short-read sequencing data in FASTQ format.**
>>- **The tool accepts both paired-end and single-end reads.**


### **Standard use:** ###

``` bash 

fastqc <forward_reads.fastq.gz>  <reverse_reads.fastq.gz> -o output_directory

```

##### **Options used in the command:** #####

- `  <forward_reads.fastq.gz>: Specifies the input file containing forward reads in FASTQ format.`
- `  <reverse_reads.fastq.gz>: Specifies the input file containing reverse reads in FASTQ format.`
- ` -o <output_directory>: Specifies the output directory where the results will be saved.`


#### **Results:** ####

 
- ` <forward_reads_fastqc>.html`
- ` <forward_reads_fastqc>.zip`
- ` <reverse_reads_fastqc>.html`
- ` <reverse_reads_fastqc>.zip`
